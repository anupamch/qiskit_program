```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
random_num = '011'
qc = QuantumCircuit(len(random_num)+1,len(random_num))
qc.h(range(len(random_num)))
qc.x(len(random_num))
qc.h(len(random_num))
qc.barrier()
for ii,yesno in enumerate(reversed(random_num)):
  if yesno == '1':  
    qc.cx(ii,len(random_num))

qc.barrier()
qc.h(range(len(random_num)))
qc.barrier()
qc.measure(range(len(random_num)),range(len(random_num)))
qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc,backend=simulator,shots=1).result()
count = result.get_counts()
print(count)
```

```python
plot_histogram(count)
```
