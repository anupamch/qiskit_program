```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector,plot_histogram
```

```python
qr = QuantumRegister(3,name='q')
crz,crx,cres = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx"),ClassicalRegister(1, name="cres")
qc = QuantumCircuit(qr,crz,crx,cres)
qc.x(0)
qc.h(1)
qc.cx(1,2)
qc.barrier()
qc.cx(0,1)
qc.h(0)
qc.barrier()
qc.measure(0,0)
qc.measure(1,1)
qc.x(2).c_if(crx,1)
qc.z(2).c_if(crz,1)
qc.draw(output='mpl')
```

```python
qr = QuantumRegister(3,name='q')
crz,crx,cres = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx"),ClassicalRegister(1, name="cres")
qc = QuantumCircuit(qr,crz,crx,cres)
qc.h(0)
qc.h(1)
qc.cx(1,2)
qc.barrier()
qc.cx(0,1)
qc.h(0)
qc.barrier()
qc.measure(0,0)
qc.measure(1,1)
qc.barrier()
qc.cx(1,2)
qc.cz(0,2)
qc.draw(output='mpl')
```

```python
qc.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc,backend = simulator,shots=1024).result()
count = result.get_counts()
plot_histogram(count)
```
